```python
import numpy as np
from   numpy import random
from   scipy import stats
import matplotlib.pyplot as plt
```

# P1

### Sample 1000 random numbers from N(0,2) and plot its histogram

```python
mu, sigma = 0, 2
nsamples = 1000
samples = random.normal(mu, sigma, nsamples)

count, bins, ignored = plt.hist(samples, 30, density = True)

plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) * np.exp( - (bins - mu)**2 / (2 * sigma**2)),
         linewidth = 2, color = 'r')
plt.show()
```

### Sample 1000 random numbers from N(0,1) by sampling from a uniform distribution and then using inv cdf

```python
uniformSamples = random.random(nsamples)
normalSamples = [stats.norm.ppf(sample) for sample in uniformSamples]

mu = 0
sigma = 1

normcount, normbins, normignored = plt.hist(normalSamples, 30, density = True)
plt.plot(normbins, 1 / (np.sqrt(2 * np.pi) * sigma) * np.exp( - (normbins - mu)**2 / (2 * sigma**2)),
         linewidth=2, color = 'r')
plt.show()
```

### Sample 1000 random numbers from N(2,10) by sampling from N(0,1)

```python
mu, sigma = 0, 1
nsamples = 1000
samples = random.normal(mu, sigma, nsamples)
muTr, sigmaTr = 2, 10
samplesTr = samples * sigmaTr + muTr

# Sample mean
print('Mean: %0.2f' % np.mean(samplesTr))
# Sample variance
print('Variance: %0.2f' % np.var(samplesTr))

normcount, normbins, normignored = plt.hist(samplesTr, 30, density = True)
plt.plot(normbins, 1 / (np.sqrt(2 * np.pi) * sigmaTr) * np.exp( - (normbins - muTr)**2 / (2 * sigmaTr**2)),
         linewidth=2, color = 'r')
plt.show()
```

### Compute pi using Monte Carlo

```python
nsamples = int(1e6)

incircle = 0
for sample in range(nsamples):
    x = -1.0 + 2.0 * random.random()
    y = -1.0 + 2.0 * random.random()
    if x ** 2.0 + y ** 2.0 <= 1.0:
        incircle += 1
        
# Estimate of pi
print('Estimate of pi from %s samples is %f' % (nsamples, 4.0 * incircle/nsamples))
```
